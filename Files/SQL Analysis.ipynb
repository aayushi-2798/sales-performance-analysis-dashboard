## Region-Wise Sales

Select Region, Round(Sum(Sales),2) As Sales
From superstore
Group By Region
Order By Sales DESC;



##### The West region records the highest sales compared to the other regions. This indicates a strong market presence and customer demand in the West region.

## Top 10 Products By sales

Select Product_Name , Round(Sum(Sales),2) As Revenue
from superstore
group by Product_Name
order by Revenue Desc
LIMIT 10 ;



##  Monthly Sales Trend

Select
Year(Order_Date) As Year,
Month(Order_Date) As Month,
Round(Sum(Sales),2) As Total_Sales
From superstore
Group By Year, Month
Order By Year, Month;



##### Monthly sales show fluctations throughout theyear , with certain months experiencing significantly higher sales compared to others.

## Top Loss-Making Products

select 
Product_Name,
Round(Sum(Sales),2) As Total_Loss
from superstore
Group By Product_Name
Order By Total_Loss Asc
LIMIT 5;



## Sales by Category Segment

Select 
Segment,
Round(Sum(Sales),2) as Sales,
Round(Sum(Profit),2) as Profit
from superstore
Group By Segment
Order By Sales desc;



##### The Consumer segment generated the highest sales and profit among all customer segments , while the Corporate segment showed stable performance. The Home Office segment contributed the lowest sales but remained profitable.

## Average Discount By Category 

Select 
Category,
Round(Sum(Discount),2) As Avg_Discount
From superstore
Group by Category;



##### The office Supplies category has the highest average discount , while Technology receives the lowest average discount among all categories.

## Profit Margin by Category

Select 
category,
round((Sum(Profit)/Sum(Sales)) * 100,2) as Profit_Margin
from superstore
group by category
order by Profit_Margin Desc;



##### Technology and Office Supplies delivered strong profit margins, while Furniture showed significantly lower profitability, indicating weaker financial performance in that category.

## Running Total Sales

Select 
Order_date,
sales,
Sum(sales) Over(
order by Order_date
) as Running_Total
from superstore
Order By Order_date;





##### The Running Total sales show a continous upward trend over time, indicating consistent accumulation of revenue throughout the business period.
